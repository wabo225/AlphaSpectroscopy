```python
import numpy as np
from matplotlib import pyplot as plt

import mca
import cal
```

```python
def data_callback(data, i):
    plt.plot(cal.E(np.arange(len(data['data']))), data['data']/data['meas_live'], label=f"{data['pressure']} torr")
    plt.scatter(data['E_p'], data['data'][data['peak_channel']]/data['meas_live'])

plt.xlabel("Energy (MeV)")
plt.ylabel("Counts / s")
mca.map_folder('datafiles', 'spectrum', data_callback)
plt.show()
```
